```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data_1,target_1,test_size=0.2)
```

```python
from sklearn.svm import SVC

model=SVC(kernel='linear')
```

```python
model.fit(train_data,train_target)
```

```python
predicted_target=model.predict(test_data)

from sklearn.metrics import mean_squared_error

# Assuming predicted_target is the result of your regression model predictions
mse = mean_squared_error(test_target, predicted_target)
print('Mean Squared Error:', mse)
```

```python
predicted_target=model.predict(test_data)

from sklearn.metrics import accuracy_score

acc=accuracy_score(test_target,predicted_target)
print('Accuracy:',acc)
```
